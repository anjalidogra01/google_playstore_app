#Cleaning the Google playstore apps (Dataframe 1)

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
missing_percentage=df.isnull().sum()/len(df)*100
missing_percentage.plot(kind='bar')
plt.xlabel('Columns')
plt.ylabel('Percentage')
plt.title('Percentage of missing values in each column')
```

```python

df.dropna(subset=["Type", "Content Rating",'Current Ver','Android Ver'], inplace=True)
df.info()
```

```python
df['Rating'].value_counts()
```

```python
bins=[0, 2, 3.5, 4.5, 5]
labels=['Poor','Average','Good','Excellent']
df["Rating_category"]=pd.cut(df['Rating'],bins=bins,labels=labels)
```

```python
df['Rating_category'].value_counts()
```

```python
df['Installs']= df['Installs'].str.replace(",",'').str.replace('+','')
df['Installs']=df['Installs'].astype(int)
```

```python
df['Installs'].value_counts()
```

```python
bins=[-1,0,10,1000,10000,100000,1000000,10000000,10000000000]
labels=['no','Very low','Low','Moderate','More than moderate','High','Very High','Top Notch']
df['Installs_category']=pd.cut(df['Installs'],bins=bins,labels=labels)
```

```python
df['Installs_category'].value_counts()
```

```python
df['Reviews'] = df['Reviews'].astype(int)
```

```python
df.rename(columns={"Price": "Price($)",'Size':'Size(Mb)'},inplace=True)
df['Price($)'] = df['Price($)'].str.replace("$","")
df['Price($)']=df['Price($)'].astype(float)
```

```python
df['Price($)'].dtype
```

```python
df.head()
```

```python
df["Size(Mb)"].unique()
```

```python
def mb_kb(s):
  try:
    if s.endswith('M'):
      return float(s[:-1])
    elif s.endswith('k'):
       return float((s[:-1])/1024)
    else:
      return np.nan
  except:
    return np.nan
```

```python
df['Size(Mb)']=df['Size(Mb)'].apply(mb_kb)
```

```python
df['Price($)'].unique()
```

```python
def title(s):
  return s.title()
```

```python
df['Category']=df['Category'].apply(title)
```

```python
df.head()
```

```python
df['Android Ver'].unique()
```

```python
def convert(y):
  try:
    if y == 'Varies with device':
      return np.nan
    elif y == '4.4W and up':
      return y.replace('W','')
    else:
      return y
  except:
      return y
```

```python
df['Android Ver']= df['Android Ver'].apply(convert)
```
